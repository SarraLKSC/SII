```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("all_cnty_yr_2015.csv")
```

```python
#Drop unnecessary attributes aka cobaye identity
dropable=["superregion2","countryname","urban","edu","female","age"]
```

```python
df.drop(columns=dropable,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
#check countries listed
countries=df['iso3'].unique()
countries
```

```python
#search for outliers among countries 
df_check=df['iso3'].value_counts()
```

```python
df_check
```

```python
# mean values by coutries 
df_by_cnty=df.groupby(by="iso3").mean()
df_by_cnty
```

```python
df_by_cnty.describe()
```

```python
df_by_cnty.info()
```

```python
drop_again=[ "v01_lowerci_95","v01_upperci_95","v02_lowerci_95","v02_upperci_95","v05_lowerci_95","v05_upperci_95","v06_lowerci_95","v06_upperci_95","v10_lowerci_95","v10_upperci_95","v15_lowerci_95","v15_upperci_95","v16_lowerci_95","v16_upperci_95","v23_lowerci_95","v23_upperci_95","v36_lowerci_95","v36_upperci_95","v41_lowerci_95","v41_upperci_95","v57_lowerci_95","v57_upperci_95"]
```

```python
df_by_cnty.drop(columns=drop_again,inplace=True)
```

```python
df_by_cnty
```

```python
#normalize values
df_by_cnty_norm=(df_by_cnty - df_by_cnty.min())/ (df_by_cnty.max()-df_by_cnty.min())
```

```python
df_by_cnty_norm
```

```python
#integer encoding for country 
dico= dict(zip(countries,range(185)))
```

```python
df_by_cnty_norm= df_by_cnty_norm.reset_index()
```

```python
df_by_cnty_norm= df_by_cnty_norm.replace({'iso3':dico})
```

```python
df_by_cnty_norm
```

```python
df_by_cnty_norm.to_csv('Base_Diet_Clean.csv')
```
