```python
import pandas as pd
```

```python
df_covid=pd.read_csv('Covid19_Clean.csv',index_col=[0])
df_diet=pd.read_csv('Base_Diet_Clean.csv',index_col=[0])
```

```python
df_covid.head(10)
```

```python
df_diet.head(10)
```

```python
df_diet.rename({"iso3":"placeCode"},axis=1,inplace=True)
```

```python
df_covid.info()
```

```python
df_covid
```

```python
df_AARN=pd.merge(df_covid,df_diet,on='placeCode')
```

```python
df_AARN
```

```python
df_AARN.to_csv('Base_merged.csv')
```

```python
df_target=df_AARN[['cumule_cases','cumule_deaths']]
```

```python
df_attribute=df_AARN.drop(columns=['cumule_cases','cumule_deaths'])
```
